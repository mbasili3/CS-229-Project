```python
def scrape_day(origin, base_url, suffix, year, month, day): 
    try: 
        driver.get(base_url + suffix + year + "-" + month + "-" + day)
        time.sleep(1.5)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source)
        weather_data_list = pd.read_html(page_source, match=".*")
        if len(weather_data_list): 
            weather_data = weather_data_list[0]
            for df in weather_data_list: 
                if len(df.columns) == 10: 
                    weather_data = df
            if not len(weather_data.columns) == 10: 
                return pd.DataFrame()
        else: 
            return pd.DataFrame()
        weather_data['origin'] = origin
        weather_data['date'] = month + "/" + day + "/" + year
        print(month + "/" + day + "/" + year)
        print("Read ", len(weather_data), " rows!")
        return weather_data
    except: 
        print("Returning empty DF")
        return pd.DataFrame()
    
```

```python
driver = webdriver.Firefox(firefox_profile=fp)
for origin, suffix in origins.items():
    print(origin)
    for month in range(1, 13): 
        l = []
        for day in range(1, monthrange(2019, month)[1]+1): 
            df = scrape_day(origin, base_url, suffix, '2019', str(month), str(day))
            l.append(df)
        print("Concatenating and saving DFs")
        try: 
            origin_df = pd.concat(l, axis=0)
            print(len(origin_df), " rows gathered.")
            origin_df.to_csv(f"/home/chosk/Documents/2020_fall/cs229/project/data_raw/weather/weather_{origin}_2019_{month}.csv", index=False)
        except Exception: 
            print("Problem!")
```
